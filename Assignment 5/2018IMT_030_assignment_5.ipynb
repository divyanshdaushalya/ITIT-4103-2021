### Importing libraries and dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df = pd.read_csv("iris.data",header=None,names=['sepal_length','sepal_width','petal_length','petal_width','class'])
```

```python
df
```

```python
df.columns
```

### Data pre-processing

```python
from sklearn.preprocessing import LabelEncoder
le_class = LabelEncoder()
df['class'] = le_class.fit_transform(df['class'])
```

```python
X = df.filter(['sepal_length','sepal_width','petal_length','petal_width'])
y = df['class']
```

### Training and Test set splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
gnb = GaussianNB()
```

```python
accuracy_li = {}
```

**Using 1 Feature at a time.**

```python
tmp_X_train = X_train.filter(['sepal_length'])
gnb.fit(tmp_X_train,y_train)
tmp_X_test = X_test.filter(['sepal_length'])
y_pred = gnb.predict(tmp_X_test)

accuracy = accuracy_score(y_pred,y_test)
accuracy_train = accuracy_score(gnb.predict(tmp_X_train),y_train)
cm = confusion_matrix(y_pred,y_test)
cm_df = pd.DataFrame(cm,index = ['SETOSA','VERSICOLR','VIRGINICA'],columns =['SETOSA','VERSICOLR','VIRGINICA'])
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.show()

print("Accuracy of the model with only ",'sepal_length'," feature = ",accuracy)

accuracy_li['sepal_length'] = accuracy
```

```python
tmp_X_train = X_train.filter(['sepal_width'])
gnb.fit(tmp_X_train,y_train)
tmp_X_test = X_test.filter(['sepal_width'])
y_pred = gnb.predict(tmp_X_test)

accuracy = accuracy_score(y_pred,y_test)
accuracy_train = accuracy_score(gnb.predict(tmp_X_train),y_train)
cm = confusion_matrix(y_pred,y_test)
cm_df = pd.DataFrame(cm,index = ['SETOSA','VERSICOLR','VIRGINICA'],columns =['SETOSA','VERSICOLR','VIRGINICA'])
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.show()

print("Accuracy of the model with only ",'sepal_width'," feature = ",accuracy)

accuracy_li['sepal_width'] = accuracy
```

```python
tmp_X_train = X_train.filter(['petal_length'])
gnb.fit(tmp_X_train,y_train)
tmp_X_test = X_test.filter(['petal_length'])
y_pred = gnb.predict(tmp_X_test)

accuracy = accuracy_score(y_pred,y_test)
accuracy_train = accuracy_score(gnb.predict(tmp_X_train),y_train)
cm = confusion_matrix(y_pred,y_test)
cm_df = pd.DataFrame(cm,index = ['SETOSA','VERSICOLR','VIRGINICA'],columns =['SETOSA','VERSICOLR','VIRGINICA'])
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.show()

print("Accuracy of the model with only ",'petal_length'," feature = ",accuracy)

accuracy_li['petal_length'] = accuracy
```

```python
tmp_X_train = X_train.filter(['petal_width'])
gnb.fit(tmp_X_train,y_train)
tmp_X_test = X_test.filter(['petal_width'])
y_pred = gnb.predict(tmp_X_test)

accuracy = accuracy_score(y_pred,y_test)
accuracy_train = accuracy_score(gnb.predict(tmp_X_train),y_train)
cm = confusion_matrix(y_pred,y_test)
cm_df = pd.DataFrame(cm,index = ['SETOSA','VERSICOLR','VIRGINICA'],columns =['SETOSA','VERSICOLR','VIRGINICA'])
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.show()

print("Accuracy of the model with only ",'petal_width'," feature = ",accuracy)

accuracy_li['petal_width'] = accuracy
```

**Using 2 Features at a time.**

```python
tmp_X_train = X_train.filter(['sepal_length','petal_length'])
gnb.fit(tmp_X_train,y_train)
tmp_X_test = X_test.filter(['sepal_length','petal_length'])
y_pred = gnb.predict(tmp_X_test)

accuracy = accuracy_score(y_pred,y_test)
accuracy_train = accuracy_score(gnb.predict(tmp_X_train),y_train)
cm = confusion_matrix(y_pred,y_test)
cm_df = pd.DataFrame(cm,index = ['SETOSA','VERSICOLR','VIRGINICA'],columns =['SETOSA','VERSICOLR','VIRGINICA'])
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.show()

print("Accuracy of the model with only ",'sepal_length, petal_length'," feature = ",accuracy)

accuracy_li['sepal_length, petal_length'] = accuracy
```

```python
tmp_X_train = X_train.filter(['sepal_width','petal_length'])
gnb.fit(tmp_X_train,y_train)
tmp_X_test = X_test.filter(['sepal_width','petal_length'])
y_pred = gnb.predict(tmp_X_test)

accuracy = accuracy_score(y_pred,y_test)
accuracy_train = accuracy_score(gnb.predict(tmp_X_train),y_train)
cm = confusion_matrix(y_pred,y_test)
cm_df = pd.DataFrame(cm,index = ['SETOSA','VERSICOLR','VIRGINICA'],columns =['SETOSA','VERSICOLR','VIRGINICA'])
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.show()

print("Accuracy of the model with only ",'sepal_width, petal_length'," feature = ",accuracy)

accuracy_li['sepal_width, petal_length'] = accuracy
```

```python
tmp_X_train = X_train.filter(['petal_width','petal_length'])
gnb.fit(tmp_X_train,y_train)
tmp_X_test = X_test.filter(['petal_width','petal_length'])
y_pred = gnb.predict(tmp_X_test)

accuracy = accuracy_score(y_pred,y_test)
accuracy_train = accuracy_score(gnb.predict(tmp_X_train),y_train)
cm = confusion_matrix(y_pred,y_test)
cm_df = pd.DataFrame(cm,index = ['SETOSA','VERSICOLR','VIRGINICA'],columns =['SETOSA','VERSICOLR','VIRGINICA'])
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.show()

print("Accuracy of the model with only ",'petal_width, petal_length'," feature = ",accuracy)

accuracy_li['petal_width, petal_length'] = accuracy
```

```python
accuracy_li_train={}
for (featureName,featureData) in X_train.iteritems():
    
    tmp_X_train = X_train.filter([featureName])
    gnb.fit(tmp_X_train,y_train)
    tmp_X_test = X_test.filter([featureName])
    y_pred = gnb.predict(tmp_X_test)
    
    accuracy = accuracy_score(y_pred,y_test)
    accuracy_train = accuracy_score(gnb.predict(tmp_X_train),y_train)
    cm = confusion_matrix(y_pred,y_test)
    cm_df = pd.DataFrame(cm,index = ['SETOSA','VERSICOLR','VIRGINICA'],columns =['SETOSA','VERSICOLR','VIRGINICA'])
    plt.figure(figsize=(5,4))
    sns.heatmap(cm_df, annot=True)
    plt.title('Confusion Matrix')
    plt.show()
    
    print("Accuracy of the model with only ",featureName," feature = ",accuracy)
    
    accuracy_li[featureName] = accuracy
    accuracy_li_train[featureName] = accuracy_train
```

**Best feature that gives maximum accuracy is petal_length**

```python
for (featureName,featureData) in X_train.iteritems():
    if featureName == 'petal_length':
        continue
    
    tmp_X_train = X_train.filter([featureName,'petal_length'])
    gnb.fit(tmp_X_train,y_train)
    tmp_X_test = X_test.filter([featureName,'petal_length'])
    y_pred = gnb.predict(tmp_X_test)
    
    accuracy = accuracy_score(y_pred,y_test)
    accuracy_train = accuracy_score(gnb.predict(tmp_X_train),y_train)
    cm = confusion_matrix(y_pred,y_test)
    cm_df = pd.DataFrame(cm,index = ['SETOSA','VERSICOLR','VIRGINICA'],columns =['SETOSA','VERSICOLR','VIRGINICA'])
    plt.figure(figsize=(5,4))
    sns.heatmap(cm_df, annot=True)
    plt.title('Confusion Matrix')
    plt.show()
    
    print("Accuracy of the model with petal_length and ",featureName," feature = ",accuracy)
    
    accuracy_li[featureName+', petal_length'] = accuracy
    accuracy_li_train[featureName+', petal_length'] = accuracy_train
```

**Accuracies Comparison**

```python
acc_df_test = pd.DataFrame(accuracy_li.items(),columns=['Features','Accuracy'])
acc_df_train = pd.DataFrame(accuracy_li_train.items(),columns=['Features','Accuracy'])
accuracy_li
```

**Plotting the accuracies and checking the best parameters.**

```python
x = np.arange(0,len(accuracy_li))
y_test = np.array(acc_df_test['Accuracy'])
y_train = np.array(acc_df_train['Accuracy'])

plt.scatter(x,y_test,color='red')
plt.plot(x,y_test,color='red',label='Testing')
plt.scatter(x,y_train,color='blue')
plt.plot(x,y_train,color='blue',label='Training')
plt.legend(loc=0)
plt.xlabel('Features-label')
plt.ylabel('Accuracy')
plt.text(7,0.9,'Feature Subset Labels')
plt.text(7, 0.6, '0 - sepal_length\n1 - sepal_width\n2 - petal_length\n3 - petal_width\n4 - petal_length, sepal_length\n5 - petal_length, sepal_width\n6 - petal_length, petal_width\n', style='italic')
```

**The best feature for the training considering the train and test set accuracy = petal_length, petal_width.**
